```python
import pandas as pd
from sklearn import linear_model
import yfinance as yf
from pandas_datareader import data as pdr
import matplotlib.pyplot as plt
# import pandas_profiling
# from pandas_profiling import ProfileReport
import matplotlib.pyplot as plt
import ta
from ta.trend import macd
from ta import momentum
from ta import trend
from ta import volatility
from ta import volume
# import talib
# from talib import ADOSC
```

```python
yf.pdr_override() # 
# download dataframe
BTC = pdr.get_data_yahoo("BTC", start="2019-01-01", end="2022-06-15")
BTC
gold = pdr.get_data_yahoo("GC=F", start="2019-01-01", end="2022-06-15")
gold
DJI = pdr.get_data_yahoo("^DJI", start="2019-01-01", end="2022-06-15")
oil = pdr.get_data_yahoo("CL=F", start="2019-01-01", end="2022-06-15")
usdeur = pdr.get_data_yahoo("EUR=X", start="2019-01-01", end="2022-06-15")
usdcny = pdr.get_data_yahoo("CNY=X", start="2019-01-01", end="2022-06-15")
```

```python
# Market cap

def get_features_14(data):
  data['MACD'] = macd(data['Close'], window_slow = 26, window_fast = 12,fillna = True)
  data['PPO'] = momentum.ppo(data['Close'], window_slow = 26, window_fast = 12,fillna = True)
  data['CCI'] = trend.cci(data['High'], data['Low'], data['Close'] ,fillna = True)
  data['ROC'] = momentum.roc(data['Close'],fillna = True)
  data['Stochastic Oscillator'] = momentum.stoch(data['High'], data['Low'], data['Close'] ,fillna = True)
  data['Williams_r'] = momentum.williams_r(data['High'], data['Low'], data['Close'] ,fillna = True)
  data['SMA_20'] =trend.sma_indicator(data['Close'], window =20, fillna =True)
  data['SMA_50'] =trend.sma_indicator(data['Close'], window =50, fillna =True)
  data['SMA_100'] =trend.sma_indicator(data['Close'], window =100, fillna =True)
  data['EMA_20'] =trend.ema_indicator(data['Close'], window =20, fillna =True)
  data['EMA_50'] =trend.ema_indicator(data['Close'], window =50, fillna =True)
  data['EMA_100'] =trend.ema_indicator(data['Close'], window =100, fillna =True)
  data['Bollinger_Bands_L'] = volatility.bollinger_lband(data['Close'], fillna =True)
  data['Bollinger_Bands_M'] = volatility.bollinger_mavg(data['Close'], fillna =True) 
  data['Bollinger_Bands_H'] = volatility.bollinger_hband(data['Close'], fillna =True)
  data['OBV'] = volume.on_balance_volume(data['Close'], data['Volume'], fillna =True)
  data['Chaikin'] = volume.chaikin_money_flow(data['High'], data['Low'], data['Close'], data['Volume'], fillna =True)
  data['MFI'] = volume.money_flow_index(data['High'], data['Low'], data['Close'], data['Volume'], fillna =True)
  data['ATR'] = volatility.average_true_range(data['High'], data['Low'], data['Close'] ,fillna = True)
  data['Open_RSI'] = momentum.rsi(data['Open'], fillna = True)
  data['Close_RSI'] = momentum.rsi(data['Close'], fillna = True)
  return data
```

```python
BTC_featues = get_features_14(BTC)
BTC_featues
```

```python
# get gold data
def get_gold_data(gold):
  gold['Movement_gold_Open'] = gold['Open'].diff(periods=1).fillna(0)
  gold['Open_RSI'] = momentum.rsi(gold['Open'], fillna = True)
  gold['Close_RSI'] = momentum.rsi(gold['Close'], fillna = True)
  gold = gold.iloc[:,6:]
  prefix = 'gold_'
  gold = gold.add_prefix(prefix)
  return gold

gold = get_gold_data(gold)
gold
```

```python
def get_data_DJI(DJI):
  DJI['Open_SMA50'] = trend.sma_indicator(DJI['Open'], window =50, fillna =True)
  DJI = DJI[['Volume', 'Open_SMA50']]
  prefix = 'DJI_'
  DJI = DJI.add_prefix(prefix)
  return DJI

DJI = get_data_DJI(DJI)
```

```python
def get_data_oil(oil):
  oil['Diff_close'] = oil['Close'].diff(periods=1).fillna(0)
  oil = oil[['Volume', 'Diff_close']]
  prefix = 'oil_'
  oil = oil.add_prefix(prefix)
  return oil

oil = get_data_oil(oil)
oil
```

```python
def get_data_usdeur(usdeur):
  usdeur['Open_SMA50'] = trend.sma_indicator(usdeur['Open'], window =50, fillna =True)
  usdeur = usdeur[['Open', 'Adj Close', 'Open_SMA50']]
  prefix = 'usdeur_'
  usdeur = usdeur.add_prefix(prefix)
  return usdeur

usdeur = get_data_usdeur(usdeur)
usdeur
```

```python
def get_data_usdcny(usdcny):
  usdcny = usdcny[['Open', 'Close', 'High']]
  prefix = 'usdcny_'
  usdcny = usdcny.add_prefix(prefix)
  
  return usdcny

usdcny = get_data_usdcny(usdcny)
usdcny
```

```python
# get BTC oval all data
def get_overall_data(data, gold, DJI, oil, usdeur, usdcny):
  final_data = data.merge(DJI, on='Date')
  final_data = final_data.merge(gold, on ='Date')
  final_data = final_data.merge(usdeur, on ='Date')
  final_data = final_data.merge(oil, on ='Date')
  final_data = final_data.merge(usdcny, on ='Date')
  return final_data

BTC_all_features = get_overall_data(BTC, gold, DJI, oil, usdeur, usdcny)

```
